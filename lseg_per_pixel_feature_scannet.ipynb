```python

def extract_lseg_img_feature(image, transform, evaluator, label=''):
    # load RGB image
    # image = Image.open(img_dir)
    # image = np.array(image)
    # image = transform(image).unsqueeze(0)
    with torch.no_grad():
        outputs = evaluator.parallel_forward(image, label)
        feat_2d = outputs[0][0].half().to("cuda:1")

    return feat_2d

feat = extract_lseg_img_feature(image, transform, evaluator).to("cuda:1")# with out label it will return the feature map of the image
feat = feat[:,:h, :w]#[512, H,W] H,W are 1/2 of the original image
```
